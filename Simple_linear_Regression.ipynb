```python
X_train1,X_test1,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(X_train1)
```

```python
model1 = LinearRegression()
model1.fit(X_train1,y_train)
```

```python
prediction = model1.predict(X_test1)
```

```python
plt.scatter(X_train1,y_train,color='red')
plt.plot(X_train1,model1.predict(X_train1),color='blue')
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test1,y_test,color='red')
plt.plot(X_train1,model1.predict(X_train1),color='blue')
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("years of experience")
plt.ylabel("Salary")
plt.show()
```
